```python
#Importing needed packages
import numpy as np
import pandas as pd
from scipy import stats
```

```python
#Reading data
Data = pd.read_csv('Sales_add (1).csv')
```

```python
Data
```

```python
# Stating hypothesis
# Null hypothesis (H0) : Average sales before digital add is equal to average sales after digital add
# Alternate hypothesis (H1) : Average sales after digital add is greater than average sales before digital add
```

```python
x = Data['Sales_before_digital_add(in $)']
```

```python
y = Data['Sales_After_digital_add(in $)']
```

```python
# Mean sales before digital add
x.mean()
```

```python
# Mean sales after digital add
y.mean()
```

```python
#Conducting indepedent t test
stats.ttest_ind(x,y)
```

```python
#Inference : Here the P-value is less than 0.05 which leads to the rejection of null hypothesis. Hence we can conclude that average sales after digital add is greater than average sales before digital add
```

```python
#Importing needed packages
import numpy as np
import pandas as pd
from scipy import stats
```

```python
#Reading data
Data = pd.read_csv('Sales_add (1).csv')
Data
```

```python
#Finding counts for category 'Region'
Data['Region'].value_counts()
```

```python
#Finding counts for category 'Manager'
Data['Manager'].value_counts()
```

```python
# Stating hypothesis
# Null hypothesis (H0) : There is no significant relationship between 'Region' and 'Manager'
# Alternate hypothesis (H1) : There is significant relationship between 'Region' and 'Manager'
```

```python
#Constructing contingency table
a = pd.crosstab(Data.Region, Data.Manager, margins=True)
a
```

```python
#Conducting independence chi-square test
obs = np.array([a.iloc[0][0:3].values,
                  a.iloc[1][0:3].values,
                a.iloc[2][0:3].values])
stats.chi2_contingency(obs)[0:3]
```

```python
#Inference : Here the P-value is greater than 0.05 which leads to the acceptance of null hypothesis.Hence we can conclude that there is no significant relationship between 'Region' and 'Manager' 
```
